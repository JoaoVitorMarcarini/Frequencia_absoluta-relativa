```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('alunos.csv')
df.head(10)
```

```python
bins = [1.50, 1.60, 1.70, 1.80, 1.90, 2.00]

labels = [
    "[1.50 - 1.60)",
    "[1.60 - 1.70)",
    "[1.70 - 1.80)",
    "[1.80 - 1.90)",
    "[1.90 - 2.00)"
]

df['Grupo Altura'] = pd.cut(df['Altura'], bins=bins, labels=labels, right=False)
df
```

```python
frequencia_altura_df = df['Grupo Altura'].value_counts().sort_index().reset_index().rename(columns={'index': 'Grupo Altura', 'count': 'Frequência Absoluta'})
frequencia_altura_df['Frequência relativa'] = (frequencia_altura_df['Frequência Absoluta'] / frequencia_altura_df['Frequência Absoluta'].sum()).apply(lambda x: f"{x:.1%}")
frequencia_altura_df
```

```python
plt.bar(x = frequencia_altura_df['Grupo Altura'], height=frequencia_altura_df['Frequência Absoluta'])
plt.show()
```
